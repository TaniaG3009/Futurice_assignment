```
# File location and type
file_location = "/FileStore/tables/gdp_data.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = "|"

# The applied options are for CSV files.
gdp_df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

permanent_table_name = "gdp_data"

gdp_df.write.format("parquet").mode("overwrite").saveAsTable(permanent_table_name)
```

```

gdp_cols_df = spark.sql("select * from gdp_data").limit(0)
gdp_cols_pd  = gdp_cols_df.toPandas()

gdp_cols_list= list(gdp_cols_pd.columns)

print ("Original list is : " + str(gdp_cols_list))
  
# using pop(0) to
# perform removal
gdp_cols_list.pop(0)
      
# Printing modified list 
print ("Modified list is : " + str(gdp_cols_list))
```

```
l=[]
separator = ','
#gdp_cols_list_updated = separator.join(gdp_cols_list).split(",")
#print(gdp_cols_list_updated)
n = len(gdp_cols_list)

for a in range(len(gdp_cols_list)):
  l.append("'{}".format(gdp_cols_list[a] + "'," + "gdp_data.`" + gdp_cols_list[a] + "`"))
print(l)
print(separator.join(l))
```

```
gdp_ict_cloud_df = spark.sql("""select gdp.country, 
                                   gdp.year, 
                                   cast(replace(gdp.gdp,',','') as int) as gdp,
                                   ict.nace_r2, 
                                   ict.geo as ict_geo, 
                                   ict.time as ict_time, 
                                   cast(ict.value as double) as ict_perc,
                                   cloud.unit as cloud_unit, 
                                   cloud.sizen_r2 as cloud_sizen_r2, 
                                   cloud.indic_is as cloud_indic_is, 
                                   cloud.geo as cloud_geo, 
                                   cloud.time as cloud_time,
                                   cast(cloud.value as double) as cloud_usage_perc
                                   from gdp_updated gdp
                                   left join
                                   ict_percentage ict
                                   on gdp.country = ict.geo
                                   and gdp.year = ict.time
                                   left join
                                   cloud_computing_service cloud
                                   on gdp.country = cloud.geo
                                   and gdp.year = cloud.time""")

gdp_ict_cloud_df.createOrReplaceTempView("gdp_ict_cld")
```

```
new_ofc_expansion_factor_df = spark.sql("""select gdp_ict_cld.country,
                                                  gdp_ict_cld.year,
                                                  gdp_ict_cld.cloud_sizen_r2, 
                                                  coalesce(gdp_ict_cld.gdp*gdp_ict_cld.ict_perc*gdp_ict_cld.cloud_usage_perc) as attractiveness_factor
                                                  from
                                                  gdp_ict_cld
                                                  where gdp_ict_cld.gdp not like '%Office%'  --eliminiating the countries which already has office                  
                                                  """)
new_ofc_expansion_factor_df.createOrReplaceTempView("new_ofc_expansion_factor")
```

```
###### The below analysis shows the top countries for a new office #########
```

```
max_expansion_factor_df = spark.sql(""" select max(final.attractiveness_factor) as expansion_determining_factor,
                                               max(final.year) as year,
                                               final.country,
                                               final.cloud_sizen_r2
                                               from
                                                new_ofc_expansion_factor final
                                               group by final.country,
                                               final.cloud_sizen_r2
                                               order by max(final.attractiveness_factor) desc
""")

max_expansion_factor_df.createOrReplaceTempView("max_expansion_factor")
```

```
final_country_list_df = spark.sql("""select *
                                     from max_expansion_factor
                                     where expansion_determining_factor is not null""")

final_country_list_df.createOrReplaceTempView("final_country_list")
```
